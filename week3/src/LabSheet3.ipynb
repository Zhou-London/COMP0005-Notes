```python
class Node:
    def __init__(self, str, sorted_str, id):
        self.str = str
        self.sorted_str = sorted_str
        self.id = id

    def __str__(self):
        return self.str


class Anagrams:
    def __init__(self, ls):
        self.ls = ls
        self.len = len(ls)

    def merge_sort_string(self, s):
        if len(s) <= 1:
            return s

        mid = len(s) // 2
        left_half = self.merge_sort_string(s[:mid])
        right_half = self.merge_sort_string(s[mid:])

        return self.merge(left_half, right_half)

    def merge(self, left, right):
        result = []
        i, j = 0, 0

        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                result.append(left[i])
                i += 1
            else:
                result.append(right[j])
                j += 1

        result.extend(left[i:])
        result.extend(right[j:])

        return "".join(result)
    
    def solve(self):
        nodeList = [Node(None, None, None)] * self.len
        for i in range(self.len):
            nodeList[i] = Node(self.ls[i], self.merge_sort_string(self.ls[i]), i)

        for i in range(self.len - 1):
            node = nodeList[i]
            if(node.id != i):
                continue
            for j in range(i+1, self.len):
                if(nodeList[j].sorted_str == node.sorted_str):
                    nodeList[j].id = node.id

        group = {}

        for node in nodeList:
            if node.id not in group:
                group[node.id] = []

            group[node.id].append(node.str)

        for strs in group.values():
            print(strs)

a = Anagrams(["hello", "ehllo", "world", "orwld", "holle", "odlow", "e", "c"])
a.solve()
```
